### **Desarrollo**

```python
import pandas as pd
import numpy as np
```

Como ya mencionamos, el problema trata sobre clasificar a los álbumes de Spotify de acuerdo a sus grado de popularidad. Nuestro problema estará enfocado en una base de datos tomada de _Kaggle_ $[7]$ que nos brinda exactamente $160000$ registros con $45$ columnas como características, donde hay una columna llamada "popularity" que nos da el grado de popularidad de este álbum. Así pues, tendremos un sistema de ecuaciones lineales de la forma

$$A\mathbf{x} = \mathbf{b}$$

Donde $A$ es la matriz que contiene las características de todos los álbumes, $\mathbf{b}$ es el vector que indica a qué clase pertenece cada álbum y $\mathbf{x}$ es el vector que deseamos hallar, el que contiene los coeficientes correspondientes a cada característica. Más específicamente, resolveremos un sistema de ecuaciones lineales como el siguiente:

$$$$
$$\begin{bmatrix} 1 & ft_{12} & ft_{13}^2 & ...  & ft_{1n}^{n-1} \\
1 & ft_{22} & ft_{23}^2 & ... & ft_{2n}^{n-1} \\
\vdots & \vdots & \vdots &\ddots & \vdots \\
1 & ft_{m2} & ft_{m3}^2 & ... & ft_{mn}^{n-1} \end{bmatrix} \begin{bmatrix} co_1 \\ co_2 \\ \vdots \\ co_n \end{bmatrix} = \begin{bmatrix} et_1 \\ et_2 \\ \vdots \\ et_m \end{bmatrix}$$

$$$$

Donde tenemos una matriz $A$ de tamaño $m \times n$ cuyos elementos $ft_{ij}$ son las caracteristicas de cada álbum, un vector $\mathbf{x}$ cuyos elementos $co_i$ son los coeficientes asociados a cada característica y un vector $\mathbf{b}$ cuyos elementos $et_i$ son las etiquetas (o clases) de cada álbum. Es posible notar que entonces obtendremos un modelo polinomial al resolver con mínimos cuadrados, donde el grado del polinomio indica la cantidad de características que tendremos en cuenta.

Lo anterior será solucionado utilizando **mínimos cuadrados**, con el fin de obtener un modelo que clasifique los datos de una nueva base de datos donde no conocemos el grado de popularidad de los álbumes. Cabe resaltar que entonces la primera base de datos sirve como "entrenamiento" del modelo, con ella obtenemos un modelo que puede clasificar nuevos álbumes, pero que evidentemente tendrá un grado de error.

Así, tendremos un conjunto de etiquetas o clases asignadas a cada género musical. Los elementos de nuestro vector $\mathbf{b}$ perteneceran a este conjunto, llamado $\mathbf{E}$:

$$1 = [0-25)$$
$$2 = [25-50)$$
$$3 = [50-75)$$
$$4 = [75-100]$$
$$$$
$$\mathbf{E} = [1,2,3,4]$$

Como ya mencionamos en nuestro enfoque partícular, el planteamiento es utilizar mínimos cuadrados para clasificar datos multi-clase. Esto lo ejecutaremos siguiendo el procedimiento propuesto por Boyd y Vandenberghe $[1]$, el cual de manera general indica que queremos obtener un modelo
$$\hat{y} = \hat{f}(x_1, ..., x_n)$$

que nos retorne la clase a la que pertenece el dato dependiendo de sus características $x_i$ y del conjunto de clases definido. Así pues, menciona que para clasificadores multi-clase, debemos realizar procesos de clasificaciones binarias para cada $k_i$ perteneciente al conjunto $\mathbf{E}$, es decir:

  Tenemos $k$ clases o etiquetas, tomamos nuestra matriz $A$ y creamos un vector $\mathbf{\bar{b}}$ a partir del vector $\mathbf{b}$, de manera que si la fila correspondiente en el vector $\mathbf{b}$ es igual a $k_i$ (clase $i$), en el vector $\mathbf{\bar{b}}$ esta fila tendrá la etiqueta $1$, si no, entonces tendrá la etiqueta $-1$ o $0$. Ahora, obtenemos el vector $\mathbf{\bar{x}}$ resolviendo por mínimos cuadrados la ecuación:

  $$A \mathbf{\bar{x}} = \mathbf{\bar{b}}$$

  Teniendo nuestro vector de coeficientes $\mathbf{\bar{x}}$ para esta clasificación binaria para cada una de las clases $k_i$, hacemos producto entre $A$ y $\mathbf{\bar{x}}$, que me retorna un vector de $n$ filas, para saber la clase a la que pertenece cada uno de los datos según el modelo, y convierto en probabilidad cada componente de este vector resultante, para así conocer qué tan probable es que cada uno de los datos pertenezca a cada clase $k_i$. La siguiente representación ayuda a entender lo enunciado:

  $$\begin{bmatrix} a_{1j} \\ a_{2j} \\ \vdots \\ a_{nj} \end{bmatrix} = \begin{bmatrix} pr_{1j.k_{1}} & pr_{1j.k_{2}} & ... & pr_{1j.k_{k}} \\ pr_{2j.k_{1}} & pr_{2j.k_{2}} & ... & pr_{2j.k_{k}} \\ \vdots & \vdots & \ddots & \vdots \\ pr_{nj.k_{1}} & pr_{nj.k_{2}} & ... & pr_{nj.k_{k}} \end{bmatrix}$$

  Finalmente, para cada dato escogemos la probabilidad mayor entre el conjunto de probabilidades para las clases. Lo realizamos con la función $\mathbf{argmax}$, que devuelve el índice del mayor valor de un conjunto. Lo anterior funciona gracias a que las clases son secuenciales y están ordenadas ascendentemente de izquierda a derecha. Este índice, entonces, será la clase asignada a nuestro dato.

A continuación, ejecutamos el algoritmo previamete implementado para una cantidad de datos pequeña, para esta base de datos real:

```python
def least_squares_multiclass_classifier(A, b, k):

  # K = cantidad de clases, para cada una se realiza un 'lstsq'
  # binario como lo propone boyd

  # Inicializar matrices de coeficientes y probabilidades
  x = np.zeros((A.shape[1], k))
  p = np.zeros((A.shape[0], k))

  # Para cada clase en el conjunto E
  for i in range(1, k + 1):
      # Crear un vector de etiquetas binarias bi (m x 1)
      # tomamos el b pasado como parámetro, hacemos una comparación
      # para que así quede solo con dos posibles valores: 1 o 0
      bi = np.where(b == i, 1, 0)

      # Resolver el sistema de ecuaciones A xi = bi usando lstsq
      # resolvemos esa clasficación binaria
      xi = np.linalg.lstsq(A, bi, rcond=None)[0]

      # Guardar el vector de coeficientes xi (n x 1) en la matriz x
      # : significa en cada fila y (i-1) porque i inicia desde 1, las
      # posiciones en python inician desde 0 (x es una matriz con las n filas
      # como características hallan y m columnas como clases hallan), es decir,
      # para cada clase tengo un vector 'xi' (que cubren todas las características - coeficientes)
      x[:, i - 1] = xi[:A.shape[1]]
      # obtenemos los primeros m coeficientes (correspondientes a las características)

      # Calcular el vector de probabilidades pi (m x 1) usando la función sigmoide
      # teniendo el vector xi para cada clase para todos los datos, hago
      # producto entre A y xi para saber la clase que dice el modelo a la cual
      # pertenece cada dato y convierto en probabilidad cada componente
      pi = 1 / (1 + np.exp(-A @ xi[:A.shape[1]]))

      # Guardar el vector de probabilidades pi en la matriz p
      # guardamos la probabilidad que tiene cada dato de que pertenezca
      # a cada una de las clases. P tiene como columnas la cantidad de clases
      # y como filas la cantidad de datos
      p[:, i - 1] = pi

  b_result = np.zeros((A.shape[0], 1))

  # Imprimir las predicciones
  for j in range(p.shape[0]):
      # argmax me devuelve el índice, estamos escogiendo para un dato (una fila)
      # el mayor valor que hay entre las 4 clases. Sumamos 1 para que ya tengamos la
      # clase a la que pertenece ya que Python indexa desde 0
      max_index = np.argmax(p[j, :]) + 1
      b_result[j] = max_index
      print(f"El álbum {j + 1} pertenece a la clase {max_index} con una probabilidad de {p[j, max_index - 1]:.2f}")

  return b_result, p, x # prueba -> devolver a x, contiene k vectores xi
  # que indican los coeficientes de cada clase para cada caracteristica
```

Explicando de manera detallada la implementación:

Tenemos una matriz de características $A$, que tiene $m$ filas y $n$ columnas. Cada fila representa un dato, y cada columna representa una característica. El vector de etiquetas $\mathbf{b}$ es la variable que deseamos predecir, sus componentes pertenecen al conjunto de clases que ya definimos.

Así, el código ejecuta el procedimiento previamente mencionado, para cada clase $k_i$ realiza el proceso de mínimos cuadrados para clasificadores binarios analizando el vector $\mathbf{b}$: si es la misma clase, asignamos un $1$, si no, asignamos un $0$. Luego, guardamos este vector de coeficientes hallado en una matriz, que luego usamos para aplicar la función Sigmoide (explicada más a detalle abajo) usando el producto punto de la matriz $A$ por el vector hallado correspondiente $\mathbf{x_i}$ (lo que da como resultado un $\mathbf{b_i}$). Así pues, esta función convierte el valor $b_i$ obtenido en una probabilidad y lo almacena en una nueva matriz de probabilidades $p$, la cual tiene como tamaño $($número de datos | $m$, número de clases | $k$$)$, que indica cuál es la probabilidad de cada dato de pertenecer a cada clase. Por último, se utilizamos la función 'argmax' para así obtener el índice de la probabilidad mayor de cada dato entre todas las clases: este resultado $+1$ (porque índices inician en $0$) será la clase.

La función **Sigmoide**, es una forma de convertir un valor numérico en una probabilidad entre $0$ y $1$. Se define como

 $$S = \frac{1}{(1 + e^{-z})}$$

donde $z$ es el valor numérico. Lo que hacemos es multiplicar la matriz de características $A$ por cada vector de coeficientes $\mathbf{x}$, y aplicar la función sigmoide al resultado. Esto nos da un vector de probabilidades para cada clase, que guardamos en una matriz $p$.

A continuación realizamos una extensión de la anterior implementación, ya que esta retorna una función (o modelo) de grado $1$, la cual nos puede brindar una buena primera aproximación. Entonces, implementamos una función que me devuelve un polinomio de grado $degree$, el cual es $\geq 1$, de acuerdo a las características de los datos, con el objetivo de obtener una posible mejor aproximación:

```python
import numpy as np

def generate_polynomial_features(A, degree):
  if degree + 1 > A.shape[1]:
    return 0
  else:
    # Creamos una nueva matriz con la primera columna llena de unos
    nueva_matriz = np.zeros((A.shape[0], degree + 1))

    # Llenamos las columnas restantes
    for i in range(0, degree+1):
      nueva_matriz[:, i] = np.ravel(A[:, i]) ** (i)

    return nueva_matriz
```

```python
def lstsq_classifier_poly(A, b, k, degree):
    # Agregar términos polinómicos a la matriz de características
    A_poly = generate_polynomial_features(A, degree)
    print('A poly: ', A_poly)
    return least_squares_multiclass_classifier(A_poly, b, k)
```

Es importante evaluar nuestro modelo en métricas que nos brinden una comprensión acerca de qué tan acertado o qué tan errado es. Para esto, podemos hacer uso de la **Matriz de Confusión**. La cual es una matriz cuadrada, donde cada columna y fila representan cada una de las clases de nuestro conjunto $\mathbf{E}$. Las columnas están de acuerdo a la clasificación que retornó el modelo y las filas de acuerdo a la clasificación real. Así, cada celda es un entero que cuenta cuántos datos coinciden con esa columna y fila. Más específicamente:

\begin{array}{cccc}
\text { Salida } & \hat{y}=1 & \hat{y}=2 & \hat{y}=3 & \hat{y}=4 & \text { Total } \\
\hline y = 1 & N_{\mathrm{t1}} & N_{\mathrm{f2}} & N_{\mathrm{f3}} & N_{\mathrm{f4}} & N_1 \\
y= 2 & N_{\mathrm{f1}} & N_{\mathrm{t2}} & N_{\mathrm{f3}} & N_{\mathrm{f4}} & N_2 \\
y = 3 & N_{\mathrm{f1}} & N_{\mathrm{f2}} & N_{\mathrm{t3}}& N_{\mathrm{f4}} & N_3\\
y = 4 & N_{\mathrm{f1}} & N_{\mathrm{f2}} & N_{\mathrm{f3}} & N_{\mathrm{t4}} & N_4\\
\text { Todo } & N_{\mathrm{t1}}+ \sum{N_{\mathrm{f1}}} & N_{\mathrm{t2}}+ \sum{N_{\mathrm{f2}}} & N_{\mathrm{t3}}+ \sum{N_{\mathrm{f3}}} & N_{\mathrm{t4}}+ \sum{N_{\mathrm{f4}}} & N
\end{array}

Se dividen todos nuestros datos en categorias. Donde:

- $N_{ti}$ es la cantidad de datos que tanto en la realidad como en el modelo, son de la clase $i$

- $N_{fi}$ es la cantidad de datos en el modelo son de la clase $i$ pero en la realidad son de otra clase

- $N_{ti} + \sum{N_{fi}}$ es la cantidad de datos que fueron clasificados por el modelo en la clase $i$ (sea en la realidad así o no)

- $N_i$ es la cantidad real de datos que pertenecen a la clase $i$

- $N$ es el Número total de datos

Varias métricas de rendimiento del conjunto de datos se expresan en términos de los números en la matriz de confusión. Es evidente que la diagonal de la matriz contiene a todos aquellas clasificaciones del modelo que son acertadas (de acuerdo a la realidad), entonces:

- **Medida de acierto (Accuracy rate):**

$$\frac{\sum_{i=1}^{k}{N_{ii}}}{N}$$

- **Medida de error (Error rate):**

$$1 - \frac{\sum_{i=1}^{k}{N_{ii}}}{N}$$

- **Ejemplo**

Para primer ejemplo de la implementación, utilizaremos una matriz $A$ relativamente pequeña, de $15\times10$, la cual simulará los álbumes y sus características. Asimismo, tendremos un vector $\mathbf{b}$ de $15\times1$, que simulará las etiquetas que posee cada álbum (del conjunto $\mathbf{E}$). Así pues, a continuación creamos estos elementos y ejecutamos el algoritmo previamente construido:

```python
# Definir la matriz 15x10
A_ex = np.zeros((15, 5))


# Asignar valores aleatorios a las columnas 1-3 dentro del rango 150000 a 400000
A_ex[:, 0:3] = np.random.uniform(150000, 400000, size=(15, 3))

# Asignar valores aleatorios al resto de columnas entre 0 y 1
A_ex[:, 3:] = np.random.uniform(0, 1, size=(15, 2))

# Crear un vector con 15 componentes entre 1 y 4 (enteros)
b_ex = np.random.randint(1, 5, size=15)

#####

# Mostrar albumes con sus clases asignadas previamente
print("\nDATOS DE ENTRADA (reales):")
df = pd.DataFrame({'Album': np.arange(1, 16), 'Vector': b_ex})
print(df)

print("\n\nDATOS OBTENIDOS POR EL MODELO:")

n = 4
b_results = []
b_results_no_poly = []

print("\n\nModelo no polinomial")
b_result_no_poly, p_result_no_poly, x_result_no_poly = least_squares_multiclass_classifier(A_ex, b_ex, 4)
b_results_no_poly.append(b_result_no_poly)

for j in range(1, n+1):
  print(f"\n\nModelo polinomial de grado {j}")
  b_result, p_result, x_result = lstsq_classifier_poly(A_ex, b_ex, 4, j)
  b_results.append(b_result)
```

Así, realizamos un gráfico comparativo para cada modelo polinomial de grado $degree$ vs los datos reales que le pasamos a la implementación, para así evaluar y conocer cuáles clasificaciones fueron acertadas o no:

Calculemos la matriz de confusión:

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(b_ex, b_results_no_poly[0])

print(f'Matriz confusión para no polinomial \n', confusion, '\n')
accuracy_rate = sum(confusion.diagonal()) / A_ex.shape[0]
print(f'Accuracy rate de ', accuracy_rate * 100, '%\n')

# Crear la matriz de confusión
n_conf = 4

for i in range(1, n_conf+1):
  confusion = confusion_matrix(b_ex, b_results[i-1])

  print(f'Matriz confusión para grado {i}\n', confusion, '\n')

  accuracy_rate = sum(confusion.diagonal()) / A_ex.shape[0]
  print(f'Accuracy rate de ', accuracy_rate * 100, '%\n')
```

Creamos la función que escoge la mejor clase para cada dato, ya que del modelo anterior (entrenamiento) obtuvimos el vector de coeficientes $\mathbf{x}$:

```python
def select_class(A, x, k):
  p_test = np.zeros((A.shape[0], k))

  # Para cada clase en el conjunto E
  for i in range(1, k + 1):
      pi = 1 / (1 + np.exp(-A @ x[:, i - 1][:A.shape[1]]))

      p_test[:, i - 1] = pi

  b_result_test = np.zeros((A.shape[0], 1))

  # Imprimir las predicciones
  for j in range(p_test.shape[0]):
      max_index = np.argmax(p_test[j, :]) + 1
      b_result_test[j] = max_index
      print(f"El álbum {j + 1} pertenece a la clase {max_index} con una probabilidad de {p_test[j, max_index - 1]:.2f}")

  return b_result_test, p_test
```

Como último paso, hacemos uso de Scikit-learn, comúnmente conocida como `sklearn`, la cual es una biblioteca de aprendizaje automático (machine learning). De manera general, en el siguiente código:
  - `LinearSVC` es el clasificador de vectores de soporte lineal que se utiliza para construir clasificadores binarios.
  - `OneVsRestClassifier` se encarga de extender LinearSVC a un problema de clasificación multiclase mediante la estrategia "one-vs-rest", entrenando un clasificador binario por cada clase.

El algoritmo me retorna un 'accuracy rate' del clasificador cuando se testea con una porción de una base de datos que fue usada para entrenar, así como hicimos previamente. Así, podremos comparar las medidas de los aciertos, las cuales son cercanas e incluso nuestro modelo es aproximadamente un $3$% mejor.

```python
# SKLEARN
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(A, b, test_size=0.2, random_state=42)

# Crear el clasificador LinearSVC con la estrategia "one-vs-rest" (OneVsRestClassifier)
classifier = OneVsRestClassifier(LinearSVC(random_state=42))

# Entrenar el clasificador
classifier.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = classifier.predict(X_test)

# Calcular la precisión del clasificador
accuracy = accuracy_score(y_test, y_pred)
print(f"\nPrecisión del clasificador: {accuracy:.2f}")
```

## **Conclusiones**
  - Los mínimos cuadrados son una técnica válida para la clasificación de datos, pero no es la más eficiente ni tampoco la más precisa. Existen estrategias más adecuadas, como las máquinas de soporte vectorial (SVM) y métodos basados en árboles de decisión

  - Un modelo polinomial no garantiza una tasa de acierto aceptable, incluso, puede llegar a ser mejor utilizar un modelo no polinomial (dependiendo del contexto). Así, encontrar el grado óptimo del polinomio puede ser un desafío: un grado muy bajo puede no capturar la complejidad de los datos, mientras que un grado muy alto puede conducir al sobreajuste.

  - Aunque mínimos cuadrados no sea la mejor opción para clasificación, su estudio en el curso es más que valioso. Proporciona una comprensión fundamental de la resolución de sistemas de ecuaciones lineales y la optimización, conceptos que son esenciales en una variedad de aplicaciones.

  - La estrategia de Boyd y Vandenberghe para clasificadores multi-clase es válida y relativamente sencilla de implementar, y aporta una solución coherente. A su vez, esta distribuye la tarea de clasificación de múltiples clases en problemas más manejables de clasificación binaria, lo que resulta en un rendimiento eficiente.
